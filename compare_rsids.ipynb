```python
import pandas as pd 

# Read and store content 
# of an excel file
read_file = pd.read_excel ("common_snps.xlsx") 
 
# Write the dataframe object 
# into csv file 
read_file.to_csv ("common_snps.csv",
                  index = None, 
                  header=True) 
    
# read csv file and convert  
# into a dataframe object 
df = pd.DataFrame(pd.read_csv("common_snps.csv")) 
  
# show the dataframe 
df
```

```python
df.loc[df["Phenotype"] == "Vitiligo"]
```

```python
import numpy as np
types = {
    "type2": ["Rheumatoid_arthritis", "Primary_biliary_cirrhosis", "Systemic_sclerosis"],
    "type3": ["Vitiligo", "Type_1_diabetes"]
}

for type in types:
    for disease in types[type]:
        rows = df[df["Phenotype"] == disease][df["Ancestry"] == "EU"][["Pos", "RS ID"]]
        print(len(rows), len(df[df["Phenotype"] == disease]))
        rows["Chr"] = [x.split(":")[0] for x in rows["Pos"]]
        rows["Start"] = [x.split(":")[1] for x in rows["Pos"]]
        rows["End"] = rows["Start"]
        del rows["Pos"]
        rows.to_csv(f"mas/{type}/{disease}.snp", "\t", header=False, index=False, columns=["Chr", "Start", "End", "RS ID"])
```
